```python
#using constructor

new_dict = dict(name="den",place="ktm",age="23")
print(new_dict)

##accesing item
x=new_dict["name"]
print(x)

#keys
print(new_dict.keys())


#values
print(new_dict.values())


#get
print(new_dict.get("age"))



#update
print(new_dict.update({"age":27}))
print(new_dict)
print(new_dict.update({"gender":"male"}))
print(new_dict)

#items

print(new_dict.items())

#copy
new2_dic = new_dict.copy()
print(new2_dic)

#pop

new2_dic.pop("age")



#popitem #remove last inserted item

new2_dic.popitem()

print(new2_dic)


#setdefault #

new2_dic.setdefault("name","den")
new2_dic.setdefault("age","24")
print(new2_dic)

new3_dic=new2_dic.copy()
print(new3_dic)


#clear

new3_dic.clear()
print(new3_dic)













```
